```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
import json
```

```python
df_master = json.load(open("hospital_stats_ny_with_costs_with_safety_with_loc.json"))
df_master
```

```python
bias_df = json.load(open("hospital_disparities.json"))
bias_df
```

```python
hospital_biases = {v["name"]: v for v in bias_df.values()}

j = 0
for hid in df_master:
    name = df_master[hid]["name"].lower()
    name = "".join([a for a in name if a in "qwertyuiopasdfghjklzxcvbnm "])
    added = False
    for hospital_name in hospital_biases:
        hn = "".join([a for a in hospital_name.lower() if a in "qwertyuiopasdfghjklzxcvbnm "])
        if name in hn or hn in name:
            df_master[hid]["racial_disparities"] = hospital_biases[hospital_name]["racial_disparities"]
            added = True

    if not added:
        df_master[hid]["racial_disparities"] = {}
        j += 1
        print("failed to find jits for", df_master[hid]["name"])
print(j)
```
